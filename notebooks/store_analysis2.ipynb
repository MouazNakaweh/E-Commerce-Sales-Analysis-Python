## 6. Trend Analysis
Monthly sales trends and state-wise sales analysis

```python
monthly = monthly_sales(df)
plt.figure(figsize=(10,6))
monthly.plot(kind='line', marker='o', color='green')
plt.title('Monthly Sales Trends')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
